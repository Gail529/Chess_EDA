# Most preffered chess openings

```
opening_list=data['opening_name'].value_counts().head(50)
opening_list.count
```

```
values=opening_list.values
names=opening_list.index
names[8]
```

```
sns.set_theme(style='darkgrid',palette="mako")
plt.figure(figsize=(10,8))
ax= data['opening_name'].value_counts().head(20).plot(kind='barh')
```

```
#least preferred openings
sns.set_theme(style='darkgrid',palette="mako")
plt.figure(figsize=(10,8))
ax= data['opening_name'].value_counts().tail(5).plot(kind='barh')
```

```


favoured_black=[]
favoured_white=[]

for n in names:
    opening_b=data.loc[(data['opening_name'] == n ) & ( data['winner'] == 'black')]
    index_b=opening_b.index
    opening_w=data.loc[(data['opening_name'] == n ) & ( data['winner'] == 'white')]
    index_w=opening_w.index
    if  len(index_b) > len(index_w):
        favoured_black.append(n)
    else :
        favoured_white.append(n)
        
```

# Openings that favoured black

```
favoured_black
```

```
opening_data=data.loc[data['opening_name'].isin(favoured_black) ]
sns.set_theme(style='darkgrid')
plt.figure(figsize=(10,8))
ax = sns.countplot(x="winner",data=opening_data,palette="mako")
```

```
black_won_data=data.loc[data['winner'] == 'black']
sns.set_theme(style='darkgrid',palette="deep")
plt.figure(figsize=(10,8))
ax= black_won_data['opening_name'].value_counts().head(20).plot(kind='barh')
```
